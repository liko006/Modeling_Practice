```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# create sample data for x^2
posx = 5 * np.random.rand(1000, 1)
negx = -5 * np.random.rand(1000, 1)
x = np.concatenate((posx, negx), axis=0)
y = x * x
```

```python
model = tf.keras.Sequential([ tf.keras.layers.Dense(units=100, activation='relu',input_shape=(1,)),
                              tf.keras.layers.Dense(units=1) ]) 

model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.01), loss='mse')
```

```python
history = model.fit(x, y, epochs=100)

x_pred = np.arange(min(x), max(x), 0.01) 
y_pred = model.predict(x_pred) 

plt.plot(x, y, 'bo') 
plt.plot(x_pred, y_pred, 'r-') 
plt.show()
```

```python
history = model.fit(xs, y, epochs=100)

xs_pred = np.arange(min(xs), max(xs), 0.01) 
y_pred = model.predict(xs_pred) 

plt.plot(xs, y, 'bo') 
plt.plot(xs_pred, y_pred, 'r-') 
plt.show()
```
